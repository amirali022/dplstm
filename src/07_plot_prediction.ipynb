##### import statements

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from datetime import datetime
```

##### read input and results

```python
df = pd.read_csv( "data/result/source_price.csv").drop( "Unnamed: 0", axis=1)
lstm_without_news = np.load( "data/result/lstm_without_news_on_sp500_denormalized_prediction_all.npy")
lstm_with_news = np.load( "data/result/lstm_with_news_on_sp500_denormalized_all.npy")
dplstm = np.load( "data/result/dp-lstm_with_news_on_sp500_denormalized_all.npy")
```

##### plot data

```python
x = [ datetime.strptime( d, "%Y-%m-%d") for d in df[ "date"].values]

plt.figure()
plt.plot( x, df[ "Adj Close"].values, label="S&P500")
plt.plot( x[ 10:], lstm_without_news, label="LSTM without news")
plt.plot( x[ 10:], lstm_with_news, label="LSTM with news")
plt.plot( x[ 10:], dplstm, label="DP-LSTM")
plt.legend()
plt.show()
```
