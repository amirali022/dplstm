##### import statements

```python
import pandas as pd
```

##### read input csv data

```python
sentiment_df = pd.read_csv( "data/result/us_financial_news_data_with_sentiment.csv", index_col="index").drop( "Unnamed: 0", axis=1)
```

```python
sentiment_df.head()
```

```python
dates = []

for i in range( len( sentiment_df)):
    dates.append( sentiment_df[ "published_date"][ i][ 0:10])

sentiment_df[ "date"] = dates

sentiment_df.head()
```

##### group data by source name

```python
sentiment_df[ "source_name"].unique()
```

###### wsj.com

```python
sentiment_df_wsj = sentiment_df[ ( sentiment_df[ "source_name"] == "wsj.com")]

print( f"count of news article from source wsj.com: { len( sentiment_df_wsj)}")

sentiment_df_wsj.head()
```

###### cnbc.com

```python
sentiment_df_cnbc = sentiment_df[ ( sentiment_df[ "source_name"] == "cnbc.com")]

print( f"count of news article from source cnbc.com: { len( sentiment_df_cnbc)}")

sentiment_df_cnbc.head()
```

###### fortune.com

```python
sentiment_df_fortune = sentiment_df[ ( sentiment_df[ "source_name"] == "fortune.com")]

print( f"count of news article from source fortune.com: { len( sentiment_df_fortune)}")

sentiment_df_fortune.head()
```

###### reuters.com

```python
sentiment_df_reuters = sentiment_df[ ( sentiment_df[ "source_name"] == "reuters.com")]

print( f"count of news article from source reuters.com: { len( sentiment_df_reuters)}")

sentiment_df_reuters.head()
```

##### group data by date

###### wsj.com

```python
sentiment_df_wsj_group_by_date = sentiment_df_wsj[ [ "date", "compound", "neg", "neu", "pos"]].groupby( "date").mean()

print( f"count of days of wsj.com: { len( sentiment_df_wsj_group_by_date)}")

sentiment_df_wsj_group_by_date.head()
```

##### cnbc.com

```python
sentiment_df_cnbc_group_by_date = sentiment_df_cnbc[ [ "date", "compound", "neg", "neu", "pos"]].groupby( "date").mean()

print( f"count of days of cnbc.com: { len( sentiment_df_cnbc_group_by_date)}")

sentiment_df_cnbc_group_by_date.head()
```

###### fortune.com

```python
sentiment_df_fortune_group_by_date = sentiment_df_fortune[ [ "date", "compound", "neg", "neu", "pos"]].groupby( "date").mean()

print( f"count of days of fortune.com: { len( sentiment_df_fortune_group_by_date)}")

sentiment_df_fortune_group_by_date.head()
```

###### reuters

```python
sentiment_df_reuters_group_by_date = sentiment_df_reuters[ [ "date", "compound", "neg", "neu", "pos"]].groupby( "date").mean()

print( f"count of days of reuters.com: { len( sentiment_df_reuters_group_by_date)}")

sentiment_df_reuters_group_by_date.head()
```

##### read S&P500 dataset

```python
sp500_df = pd.read_csv( "data/input/SP500.csv")
```

```python
sp500_df.head()
```

##### merge sp500 anb wsj

```python
sp500_merge_wsj_df = pd.merge( sentiment_df_wsj_group_by_date, sp500_df, how="right", left_on="date", right_on="Date").fillna( 0)

sp500_merge_wsj_df.to_csv( "data/result/wsj.csv")
```

merge sp500 and cnbc

```python
sp500_merge_cnbc_df = pd.merge( sentiment_df_cnbc_group_by_date, sp500_df, how="right", left_on="date", right_on="Date").fillna( 0)

sp500_merge_cnbc_df.to_csv( "data/result/cnbc.csv")
```

merge sp500 and fortune

```python
sp500_merge_fortune_df = pd.merge( sentiment_df_fortune_group_by_date, sp500_df, how="right", left_on="date", right_on="Date").fillna( 0)

sp500_merge_fortune_df.to_csv( "data/result/fortune.csv")
```

merge sp500 and reuters

```python
sp500_merge_reuters_df = pd.merge( sentiment_df_reuters_group_by_date, sp500_df, how="right", left_on="date", right_on="Date").fillna( 0)

sp500_merge_reuters_df.to_csv( "data/result/reuters.csv")
```

##### concat all datesets

```python
sp500_merge_wsj_df.rename( { "compound": "wsj_mean_compound"}, axis=1, inplace=True)
sp500_merge_cnbc_df.rename( { "compound": "cnbc_mean_compound"}, axis=1, inplace=True)
sp500_merge_fortune_df.rename( { "compound": "fortune_mean_compound"}, axis=1, inplace=True)
sp500_merge_reuters_df.rename( { "compound": "reuters_mean_compound"}, axis=1, inplace=True)
sp500_df.rename( { "Date": "date"}, axis=1, inplace=True)
```

```python
news_sentiment_with_price = pd.concat( [
    sp500_df[ [ "date", "Adj Close"]],
    sp500_merge_wsj_df[ "wsj_mean_compound"],
    sp500_merge_cnbc_df[ "cnbc_mean_compound"],
    sp500_merge_fortune_df[ "fortune_mean_compound"],
    sp500_merge_reuters_df[ "reuters_mean_compound"],
], axis=1)

news_sentiment_with_price.head()
```

```python
news_sentiment_with_price.to_csv( "data/result/source_price.csv")
```
